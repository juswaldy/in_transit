```python
import skimage
from skimage.morphology import skeletonize
from skimage import data
import matplotlib.pyplot as plt
from skimage.util import invert
```

```python
orig = skimage.io.imread('1.cropped/john0002.jpg')
inverted = invert(orig)

image = skimage.color.rgb2gray(orig)
skeleton = skeletonize(image)

inverted = skimage.color.rgb2gray(inverted)
skeleton_inv = skeletonize(inverted)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(40, 20),
                         sharex=True, sharey=True)

ax = axes.ravel()

ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('gray', fontsize=20)

ax[2].imshow(skeleton, cmap=plt.cm.gray)
ax[2].axis('off')
ax[2].set_title('skeleton', fontsize=20)

ax[1].imshow(inverted, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('inverted', fontsize=20)

ax[3].imshow(skeleton_inv, cmap=plt.cm.gray)
ax[3].axis('off')
ax[3].set_title('skeleton_inv', fontsize=20)

fig.tight_layout()
plt.show()
```
