```python
import numpy as np
import pandas as pd
import PIL
from PIL import Image
from PIL import ImageFilter
from PIL import ImageStat
from PIL import ImageOps
from glob import glob
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
files = glob("*.[Jj][Pp][Gg]")
```

```python
plt.figure()
h = []
for f in sorted(files):
    x = Image.open(f)
    plt.plot(x.histogram())
    h.append(x.histogram())

plt.show()
np.shape(h)
```

```python
plt.plot(np.mean(h, 0))
plt.show()
```

```python
x = Image.open(files[0])
x
```

```python
poster = ImageOps.posterize(x, 1)
poster
```

```python
n = np.array(poster)
n[n>0] = 255
np.shape(n)
```

```python
color = Image.fromarray(n)
color
plt.plot(color.histogram())
plt.show()
```

```python
bw = n[:,:,0]
bw = Image.fromarray(bw)
bw
```
