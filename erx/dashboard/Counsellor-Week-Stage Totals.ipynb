## Get historical data < 2017

```python
historical = df.loc[df['Year'] < 2017]
historical.describe()
```

## Get 2017 data

```python
yr2017 = df.loc[df['Year'] == 2017]
yr2017.describe()
```

## 1. Historic Averages

### a. Cumulative sum

```python
# Group by year/week.
x = historical.fillna(0).groupby(['Year', 'Week'])

# Get sum for each week.
hist_total = x.agg('sum')
hist_total.sample(7)
```

```python
# Reset index.
x = hist_total.reset_index(level=[0,1])

# Pivot on Week.
hist_cumulative = x.pivot(index='Week', columns='Year').cumsum()
hist_cumulative[:7]
```

### b. Average cumulative data

```python
# Unstack previous pivot and reset index.
x = hist_cumulative.unstack().reset_index()

# Rename columns.
x.columns = ['Stage', 'Year', 'Week', 'Cumulative']

# Drop the Year column.
x = x.drop('Year', 1)

# Get the means.
hist_averages = x.groupby(['Stage', 'Week']).mean()
hist_averages[:77]
```

## 2. Weekly Goal

### a. Diff = current annual goals - average week 52 historical amount

```python
# Reset index.
x = hist_averages.reset_index()

# Get week-52 historical amounts.
x = x.loc[x['Week'] == 52].reset_index()
week52 = x[['Stage', 'Cumulative']]
week52.columns = ['Stage', 'Week52']
week52
```

```python
# Load goals.
x = pd.read_csv('Goals.csv').sum().drop('LastName').reset_index()
goals = pd.DataFrame(x)
goals.columns =['Stage', 'Week52Goal']
goals
```

```python
# Join and make diff.
diff = week52.set_index('Stage').join(goals.set_index('Stage'))
diff['Diff'] = diff.Week52Goal - diff.Week52
diff
```

### b. Weekly goal = historic average + diff * n/52, n = current week number

```python
# Combine historic and goals.
x = hist_averages.reset_index()
y = diff.reset_index()
hist_n_goals = x.set_index('Stage').join(y.set_index('Stage'))
hist_n_goals.columns = ['Week', 'Historical_Cumulative', 'Week52', 'Week52Goal', 'Diff']
hist_n_goals.sample(5)
```

```python
# Calculate weekly goals.
wgoal = hist_n_goals
wgoal['WeeklyGoal'] = wgoal.Historical_Cumulative + wgoal.Diff * wgoal.Week / 52
wgoal = wgoal.drop(['Week52', 'Week52Goal', 'Diff'], 1)
wgoal[:7]
```

## 3. Goal Variance = current data – weekly goal

```python
# Prepare current data.
x = yr2017.drop('Year', 1)

# Remove Alexander and Bustard.
x = x.fillna(0).loc[~yr2017['LastName'].isin(['Alexander', 'Bustard'])]

# Get cumulative sum by week.
x = x.groupby('Week').agg('sum').cumsum()

# Flatten the structure.
currentdata = x.unstack().reset_index()
currentdata.columns = ['Stage', 'Week', 'Current_Cumulative']
currentdata[:7]
```

```python
# Prepare final df.
result = wgoal.reset_index().set_index(['Week', 'Stage']).join(currentdata.reset_index().set_index(['Week', 'Stage']))

# Get goal variance.
result['GoalVariance'] = result.Current_Cumulative - result.WeeklyGoal
result = result.drop(['index'], 1).reset_index()
#result = result.pivot(index='Week', columns='Stage')
result[:7]
```
