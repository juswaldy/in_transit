# Counsellor-Week-Stage Details

## Get historical data < 2017

```python
historical = df.loc[df['Year'] < 2017]
#historical.loc[historical.LastName == 'BeugelinkTNUB', 'LastName'] = 'Beugelink'
historical.describe()
```

## Get 2017 data

```python
yr2017 = df.loc[df['Year'] == 2017]
yr2017.loc[yr2017.LastName == 'Beugelink', 'LastName'] = 'BeugelinkTNUB'
yr2017.describe()
```

## 1. Historic Averages

### a. Cumulative sum

```python
# Group by year/week.
x = historical.fillna(0).groupby(['Year', 'Week', 'LastName'])

# Get sum for each week.
hist_total = x.agg('sum')
hist_total[:7]
```

```python
# Reset index.
x = hist_total.stack().reset_index()
x.columns = ['Year', 'Week', 'LastName', 'Stage', 'Count']

# Pivot on Week.
hist_cumulative = x.pivot_table(index='Week', columns=['Year', 'LastName', 'Stage']).cumsum()
hist_cumulative[:7]
```

### b. Average cumulative data

```python
# Unstack previous pivot and reset index.
x = hist_cumulative.unstack().reset_index()

# Rename columns.
x.columns = ['Measure', 'Year', 'LastName', 'Stage', 'Week', 'Historical_Cumulative']

# Drop the Year column.
x = x.drop('Measure', 1)

# Get the means.
hist_averages = x.groupby(['LastName', 'Stage', 'Week']).mean()
hist_averages = hist_averages.drop('Year', 1)
hist_averages.sample(7)
```

## 2. Weekly Goal

### a. Diff = current annual goals - average week 52 historical amount

```python
# Reset index.
x = hist_averages.reset_index()

# Get week-52 historical amounts.
x = x.loc[x['Week'] == 52].reset_index()
week52 = x[['LastName', 'Stage', 'Historical_Cumulative']]
week52.columns = ['LastName', 'Stage', 'Week52']
week52.sample(7)
```

```python
# Load goals.
x = pd.read_csv('Goals.csv')
goals = pd.DataFrame(x)
goals = goals.set_index('LastName').stack().reset_index()
goals.columns = ['LastName', 'Stage', 'Week52Goal']
goals.sample(7)
```

```python
# Join and make diff.
diff = week52.set_index(['LastName', 'Stage']).join(goals.set_index(['LastName', 'Stage']))
diff['Diff'] = diff.Week52Goal - diff.Week52
diff.sample(7)
```

### b. Weekly goal = historic average + diff * n/52, n = current week number

```python
# Combine historic and goals.
x = hist_averages.reset_index()
y = diff.reset_index()
hist_n_goals = x.set_index(['LastName', 'Stage']).join(y.set_index(['LastName', 'Stage']))
hist_n_goals.sample(5)
```

```python
# Calculate weekly goals.
wgoal = hist_n_goals
wgoal['WeeklyGoal'] = wgoal.Historical_Cumulative + wgoal.Diff * wgoal.Week / 52
wgoal = wgoal.drop(['Week52', 'Week52Goal', 'Diff'], 1)
wgoal.sample(7)
```

## 3. Goal Variance = current data – weekly goal

```python
# Prepare current data.
x = yr2017.drop('Year', 1)

# Remove Alexander and Bustard.
x = x.fillna(0).loc[~yr2017['LastName'].isin(['Alexander', 'Bustard'])]

# Get cumulative sum.
x = x.pivot_table(index='Week', columns='LastName').cumsum()

# Flatten.
currentdata = x.unstack().reset_index()
currentdata.columns = ['Stage', 'LastName', 'Week', 'Current_Cumulative']
currentdata.sample(7)
```

```python
# Prepare final df.
result = wgoal.reset_index().set_index(['LastName', 'Week', 'Stage']).join(currentdata.reset_index().set_index(['LastName', 'Week', 'Stage']))

# Get goal variance.
result['GoalVariance'] = result.Current_Cumulative - result.WeeklyGoal
result = result.drop(['index'], 1).reset_index()
result.sample(7)
```
