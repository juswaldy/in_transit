```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi

import skimage
from skimage import io, data, feature, measure
from PIL import Image
```

```python
im = skimage.io.imread('cut/05.1.1.jpg', as_grey=True)
im.shape
```

```python
im = Image.open('cut/05.1.1.jpg')
mono = im.convert('L')
mono = np.array(mono)
im = mono
plt.imshow(im)
```

```python
# Find contours at a constant value of 0.8
contours = measure.find_contours(im, 0.8)

# Display the image and plot all contours found
fig, ax = plt.subplots()
ax.imshow(mono, interpolation='nearest', cmap=plt.cm.gray)

for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

ax.axis('image')
plt.show()
```
