```python
from PIL import Image
width = 512
height = 512
params = "-s5.00-c0.05-t1.00-"
```

```python
def loadandresize(width, height, filename):
    f = Image.open(filename)
    fsize = f.size
    ratio = max(width/fsize[0], height/fsize[1])
    resultsize = (int(fsize[0]*ratio), int(fsize[1]*ratio))
    return f.resize(resultsize)
```

## 1. The base image

```python
basename = 'board1'
loadandresize(width, height, "styletransfer/pics/" + basename + '.jpg')
```

## 2. The "style" image

```python
stylename = "mira/MiraButterfly.jpg"
loadandresize(width, height, "styletransfer/style/" + stylename)
```

## 3. Apply "style" to the base image

```python
filename = "styletransfer/dsgiitr/results/" + stylename + "/" + basename + params + "i010.jpg"
loadandresize(width, height, filename)
```
